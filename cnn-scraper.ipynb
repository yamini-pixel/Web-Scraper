```python
from urllib.request import Request, urlopen
import html5lib
from bs4 import BeautifulSoup
import requests
import csv
```

```python
topic = ['https://www.cnn.com/specials/space-science', 'https://www.cnn.com/specials/us/energy-and-environment']

with open('news-cnn.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Text'])

    for t in topic:
        response = requests.get(t)
        soup = BeautifulSoup(response.text, 'html.parser')

        # desired section for links in the main search page
        sec = soup.find('div', class_="column zn__column--idx-0")
        urls = []

        # number of links inside this part
        links = sec.select('div.media a')
        print(f'Number of links found: {len(links)}')

        # Find all anchor tags 
        links = sec.select('div.media a')

        # Loop over the anchor tags and extract the URLs
        for link in links:
            href = link.get('href')
            if href.startswith('http'):
                urls.append(href)
            else:
                urls.append('https://www.cnn.com' + href)
        print(urls)
#       Send a GET request to the URL and write data to the CSV file
        for u in urls:
            response = requests.get(u)
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find the article content
            article = soup.find('div', class_="article__content-container")
            if article is None:
                # Skip this URL if article is not found
                continue
            # Find the article title
            title = soup.find('h1').text
            text = ''
            para_list = article.find_all('p', class_="paragraph inline-placeholder")
            # Loop over the div elements and extract the information you need
            for para in para_list:
                # Get the text inside the div element
                text += para.text.strip() + ' '
            # Write data to the CSV file
            writer.writerow([title, text])
```
